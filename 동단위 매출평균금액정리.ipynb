```python
# 경로만 따로 설정해주면 될듯

df = pd.read_excel('C:\\Users\\remig\\OneDrive\\바탕 화면\\BC카드 공모전\\금융빅데이터공모전_데이터\\비씨카드_편의점만2.xlsx')
```

```python
df_new = df[df['업종명'].isin(['편 의 점'])]

# 10대 남,여 매출데이터
df_new_10 = df_new[df_new['연령대'].isin(['10대'])]
df_new_10_m = df_new_10[df_new_10['성별'].isin(['남'])]
df_new_10_f = df_new_10[df_new_10['성별'].isin(['여'])]

# 20대 남,여 매출데이터
df_new_20 = df_new[df_new['연령대'].isin(['20대'])]
df_new_20_m = df_new_20[df_new_20['성별'].isin(['남'])]
df_new_20_f = df_new_20[df_new_20['성별'].isin(['여'])]

# 30대 남,여 매출데이터
df_new_30 = df_new[df_new['연령대'].isin(['30대'])]
df_new_30_m = df_new_30[df_new_30['성별'].isin(['남'])]
df_new_30_f = df_new_30[df_new_30['성별'].isin(['여'])]

# 40대 남,여 매출데이터
df_new_40 = df_new[df_new['연령대'].isin(['40대'])]
df_new_40_m = df_new_40[df_new_40['성별'].isin(['남'])]
df_new_40_f = df_new_40[df_new_40['성별'].isin(['여'])]

# 50대 남,여 매출데이터
df_new_50 = df_new[df_new['연령대'].isin(['50대'])]
df_new_50_m = df_new_50[df_new_50['성별'].isin(['남'])]
df_new_50_f = df_new_50[df_new_50['성별'].isin(['여'])]

# 60대 이상 남,여 매출데이터
df_new_60 = df_new[df_new['연령대'].isin(['60대 이상'])]
df_new_60_m = df_new_60[df_new_60['성별'].isin(['남'])]
df_new_60_f = df_new_60[df_new_60['성별'].isin(['여'])]
```
